```python
# importar librerias

import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import csv
from pandas import ExcelWriter
```

```python
# extración población por País

website = requests.get('https://indepaz.org.co/informe-de-masacres-en-colombia-durante-el-2020-2021/')
```

# Masacres

```python
# Codigo

soup = BeautifulSoup(website.text, 'html.parser')

first_table = soup.select('.has-background')[0]

first_table
```

```python
table_row = first_table.select('tr')[1:]

table_row
```

```python
csv_data = [['No','fecha','No. de victimas','Departamento','Municipio']]
```

```python
for row in table_row[0:]:
    
    
    table_data = row.select('td')
    No = table_data[0].text
    Fecha = table_data[1].text
    Numero_victimas = table_data[2].text
    Departamento = table_data[3].text
    Municipio = table_data[4].text
    

    
    csv_data.append([No, Fecha, Numero_victimas, Departamento, Municipio ])
```

```python
csv_data
```

```python
masacres = pd.DataFrame(csv_data, columns=csv_data[0])
masacres=masacres.drop(masacres.index[[0]])
masacres['cantidad'] = 1

masacres
```

```python
# Depuración departamentos

masacres['Departamento'].replace(
    to_replace=['BOGOTA'],
    value='BOGOTÁ',
    inplace=True
)


masacres['Departamento'].replace(
    to_replace=['BOLIVAR'],
    value='BOLÍVAR',
    inplace=True
)

masacres['Departamento'].replace(
    to_replace=['CALDAS'],
    value='CÁLDAS',
    inplace=True
)

masacres['Departamento'].replace(
    to_replace=['CAQUETA'],
    value='CAQUETÁ',
    inplace=True
)

masacres['Departamento'].replace(
    to_replace=['CHOCO'],
    value='CHOCÓ',
    inplace=True
)

masacres['Departamento'].replace(
    to_replace=['CORDOBA'],
    value='CÓRDOBA',
    inplace=True
)

masacres['Departamento'].replace(
    to_replace=['GUAJIRA'],
    value='LA GUAJIRA',
    inplace=True
)

masacres['Departamento'].replace(
    to_replace=['GUAJIRA'],
    value='LA GUAJIRA',
    inplace=True
)

masacres['Departamento'].replace(
    to_replace=['VALLE DELCAUCA'],
    value='VALLE DEL CAUCA',
    inplace=True
)
```

```python
# Depuración municipios

# Depuración departamentos

masacres['Municipio'].replace(
    to_replace=['BOGOTA'],
    value='BOGOTÁ',
    inplace=True
)

masacres['Municipio'].replace(
    to_replace=['ALTO BAUDO'],
    value='ALTO BAUDÓ',
    inplace=True
)

masacres['Municipio'].replace(
    to_replace=['CAJIBIO'],
    value='CAJIBÍO',
    inplace=True
)


masacres['Municipio'].replace(
    to_replace=['CI´ÉNAGA'],
    value='CIÉNAGA',
    inplace=True
)
```

```python
masacres['Departamento'] = masacres['Departamento'].str.strip()

masacres['Departamento'] = masacres['Departamento'].str.upper()

masacres['Municipio'] = masacres['Municipio'].str.strip()

masacres['Municipio'] = masacres['Municipio'].str.upper()
```

```python
## convert str to int

masacres['No. de victimas'].replace(
    to_replace=['por definir '],
    value='0',
    inplace=True
)


masacres['No. de victimas'] = masacres['No. de victimas'].astype(int)
```

```python


departamentos = masacres.groupby(['Departamento']).sum().reset_index()

#departamentos = pd.concat([departamentos['Departamento'], departamentos['No']],axis=1)

#departamentos.columns = ['Departamento', 'No']

departamentos
```

```python
departamentos_municipio = masacres.groupby(['Departamento', 'Municipio']).sum().reset_index()

#departamentos_municipio = pd.concat([departamentos_municipio['Departamento'], departamentos_municipio['Municipio'], departamentos_municipio['No']],axis=1)

#departamentos_municipio.columns = ['Departamento', 'Municipio','No']

departamentos_municipio
```

```python
# Masacres

with ExcelWriter('Masacres_2022.xlsx') as writer:
    masacres.to_excel(writer, 'Lista_completa')
    departamentos.to_excel(writer, 'total departamento')
    departamentos_municipio.to_excel(writer, 'total municipio')
```
